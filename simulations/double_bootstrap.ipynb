```python
import numpy as np
from scipy import stats
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
sample_size = 100
n_bootstraps = 100
n_double_bootstraps = 50
samples = np.random.normal(0, 1, sample_size)
double_bootstrapped_samples = np.zeros((n_bootstraps, n_double_bootstraps, sample_size))

bootstrapped_samples = np.vstack([np.random.choice(samples, sample_size) for _ in range(n_bootstraps)])

for i in range(n_bootstraps):
    double_bootstrapped_samples[i] = np.vstack([np.random.choice(bootstrapped_samples[i], sample_size) for _ in range(n_double_bootstraps)])
```

```python
double_std = double_bootstrapped_samples.std(ddof=1, axis=2).std(ddof=1, axis=1)
```

```python
_ = plt.hist(double_std)
```

```python
bootstrapped_cohorts_std = bootstrapped_samples.std(ddof=1, axis=1)
```

```python
_ = plt.hist(bootstrapped_cohorts_std)
```

The deviation will decrease by increasing sample size (obviously!)

n=30 --> the first distribution range: [0.025-0.25] second distribtuion:[0.6-1.4]
n=100 --> the first distribution range: [0.05-0.09] second distribtuion:[0.9-1.25]
