```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
combinations_df = pd.read_csv("./simulation_combinations.csv")
```

```python
for i in range(1,7):

    df = pd.read_csv(f"../../../normal_simulation/simulation_result_{i}.csv")
    combo = combinations_df.iloc[i]
    df.p_metric.hist(alpha=0.5)
    df.pvalue.hist(alpha=0.5)
    plt.title(f"sample_size:{combo['size']}  effect_size:{combo['locations']}\
    n_bootstrap:{combo['n_bootstrap']}  n_permutation:{combo['n_permutation']}  n_simulation:{len(df)}")
    plt.legend(["p_metric", "p_value"])
    plt.show()
        
```
